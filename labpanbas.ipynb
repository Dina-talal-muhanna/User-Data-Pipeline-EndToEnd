```python
df['Is_Excellent'] = df['Grade'] >= 90
df[['Name', "Grade","Is_Excellent"]]
```

```python
df['Grade'] = df['Grade'].fillna(df['Grade'].mean())
df[['Name', 'Grade']]
```

```python
common_date = df['Score Date'].mode()[0]
df['Score Date'] = df['Score Date'].fillna(common_date)
print(df.isnull().sum())
```

```python
df['Final_Grade'] = df['Grade'] + 5
df[['Name', "Grade","Final_Grade"]]
```

```python
bins = [-1,69,79,89,100]
grades = ['Fail', 'Average', 'Good', 'Excellent']
df['Grade_Category']= pd.cut(df["Grade"], bins=bins, labels=grades)
df[['Name', "Grade","Grade_Category"]]
```

```python
df.drop(columns=['Student ID'], inplace= True)
df
```

```python
df = df[df['Passed'] == True]
df
```

```python
df = df.sort_values(by='Grade', ascending=False)
df
```

```python
df= df.sort_values(by=['City', 'Grade'], ascending=[True, False])
df
```

```python
df['Rank'] = df['Grade'].rank(ascending=False, method='min')
df
```

```python
df['avg_City'] = df.groupby('City')['Grade'].transform('mean')
df[['Name', 'City', 'Grade','avg_City']]
```

```python
sub = df.groupby("Subject")['Grade'].mean()
highst_sub_name = sub.idxmax()
highst_sub_grade = sub.max()
print(highst_sub_name, " = ", highst_sub_grade)
```
