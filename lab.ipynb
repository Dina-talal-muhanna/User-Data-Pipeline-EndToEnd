```python
data = np.array([150, 220, 90, 30, 180, 45])
min_data = data.min()
max_data = data.max()
nor = (data - min_data) / (max_data - min_data)
print(nor)
```

```python
data = np.array([150, 220, 90, 30, 180, 45])
data2 = np.mean(data)
std = np.std(data)
standardized_data = (data - data2) / std
print(standardized_data)
```

```python
mat = np.array([
    [4, 8, 1, 12],
    [7, 3, 9, 5],
    [11, 2, 6, 10],
    [15, 0, 14, 13]
])
#Compute the mean of each row
meanOfRow = np.mean(mat, axis=1)
print(meanOfRow)
print('-----------------------------------------------')
#b. Compute the sum of each column
SumOfCol = np.sum(mat, axis=0)
print(SumOfCol)
print('-----------------------------------------------')
#c. Maximum value and its position
maxval = np.max(mat) 
print(maxval)
maxpos = np.unravel_index(np.argmax(mat), mat.shape) 
print(maxpos)
```

```python
#Reshape the array np.arange(36) into a 3D array of shape (3, 4, 3).
arrThree = np.arange(36)
print(arrThree)
arrThree = np.arange(36).reshape(3, 4, 3)
print(arrThree)
```

```python
#Reverse the order of rows and columns (flip both).
arr = np.arange(1, 17).reshape(4, 4)
print(arr)
print('-----------------------------------------------')
flipped_arr = arr[::-1, ::-1]
print(flipped_arr)
```

```python
ages = np.array([23, 45, 19, 34, 28, 51, 17, 39])
names = np.array(["Ali", "Sara", "Omar", "Lina", "Khaled", "Nour", "Yara", "Hassan"])
age25to40 = (ages >= 25) & (ages <= 40)
result = names[age25to40]
print(result)
```
